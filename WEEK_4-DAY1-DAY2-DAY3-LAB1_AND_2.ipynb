### 3 - Now define a function that takes a list of models and train (and tests)

```python
from sklearn.neighbors import  KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
def model(X_train, X_test, y_train, y_test):
    # Linear model
    lm = linear_model.LinearRegression()
    model = lm.fit(X_train,y_train)
    Linear_Regression_score = lm.score(X_test,y_test)
    print('Linear_Regression_score : ')
    # KNeighborsRegressor model
    KNN = KNeighborsRegressor(n_neighbors=3)
    KNN.fit(X_train, y_train)
    KNNRegressor_score = KNN.score(X_test, y_test)
    print('KNNRegressor_score :')
    # MLPRegressor
    regr = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
    regr.predict(X_test[:2])
    MLPRegressor_score = regr.score(X_test, y_test)
    print('MLPRegressor_score :')
    return Linear_Regression_score, KNNRegressor_score, MLPRegressor_score
```

```python
model(X_train, X_test, y_train, y_test)
```
